```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
import os
os.listdir()
```

```python
df = pd.read_csv("New Year Sales Data.csv", encoding='unicode_escape')
df.head()
```

```python
df.drop(['Status', 'unnamed1'], axis=1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype(int)
```

```python
ax = sns.countplot(x='Gender', data=df)

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Count of Customers by Gender")
plt.show()
```

```python
sales_gender = df.groupby('Gender', as_index=False)['Amount'].sum()

sales_gender
```

```python
ax = sns.countplot(
    data=df,
    x='Age Group',
    hue='Gender'
)

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Age Group Distribution by Gender")
plt.show()
```

```python
age_sales = df.groupby(
    ['Age Group'],
    as_index=False
)['Amount'].sum()

age_sales.sort_values(
    by='Amount',
    ascending=False
)
```

```python
state_orders = df.groupby(
    ['State'],
    as_index=False
)['Orders'].sum().sort_values(
    by='Orders',
    ascending=False
).head(10)

state_orders
```

```python
state_amount = df.groupby(
    ['State'],
    as_index=False
)['Amount'].sum().sort_values(
    by='Amount',
    ascending=False
).head(10)

state_amount
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    data=state_orders,
    x='State',
    y='Orders'
)

plt.xticks(rotation=45)
plt.title("Top 10 States by Orders")
plt.show()
```

```python
plt.figure(figsize=(12,5))

sns.barplot(
    data=state_amount,
    x='State',
    y='Amount'
)

plt.xticks(rotation=45)
plt.title("Top 10 States by Revenue")
plt.show()
```

```python
ax = sns.countplot(
    x='Marital_Status',
    data=df
)

for bars in ax.containers:
    ax.bar_label(bars)

plt.title("Customer Count by Marital Status")
plt.show()
```

```python
marital_sales = df.groupby(
    ['Marital_Status','Gender'],
    as_index=False
)['Amount'].sum()

marital_sales
```

```python
occupation_sales = df.groupby(
    ['Occupation'],
    as_index=False
)['Amount'].sum().sort_values(
    by='Amount',
    ascending=False
)

occupation_sales.head(10)
```

```python
plt.figure(figsize=(15,5))

sns.barplot(
    data=occupation_sales.head(10),
    x='Occupation',
    y='Amount'
)

plt.xticks(rotation=45)
plt.title("Top Occupations by Revenue")
plt.show()
```

```python
category_sales = df.groupby(
    ['Product_Category'],
    as_index=False
)['Amount'].sum().sort_values(
    by='Amount',
    ascending=False
)

category_sales.head(10)
```

```python
plt.figure(figsize=(15,5))

sns.barplot(
    data=category_sales.head(10),
    x='Product_Category',
    y='Amount'
)

plt.xticks(rotation=45)
plt.title("Top Product Categories by Revenue")
plt.show()
```
